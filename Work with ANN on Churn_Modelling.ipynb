# Start work with neural network

### Importing the libraries

```python
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.preprocessing import LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, accuracy_score, classification_report
from sklearn.neural_network import MLPRegressor, MLPClassifier
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
```

## Part 1 - Data Preprocessing

### Importing the dataset

```python
dataset = pd.read_csv('/content/Churn_Modelling.csv')
dataset
```

```python
X = dataset.drop(["Exited","RowNumber","CustomerId","Surname"], axis=1)
y = dataset["Exited"]
```

```python
X
```

```python
y
```

### Encoding categorical data

Label Encoding the "Gender" column

```python
le = LabelEncoder()
X['Gender'] = le.fit_transform(X['Gender'])
```

```python
X
```

One Hot Encoding the "Geography" column

```python
X = pd.get_dummies(X, columns=['Geography'], drop_first=False ).astype(int)
```

```python
X
```

Feature Scaling

```python
sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
X
```

### Splitting the dataset into the Training set and Test set

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

## Part 2 - Start for use Nneural network

### MLP Regression

```python
mlp_reg = MLPRegressor(hidden_layer_sizes=(64, 32), activation='relu', solver='adam', max_iter=1000, random_state=42)
mlp_reg.fit(X_train, y_train)
```

```python
y_pred_reg = mlp_reg.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred_reg)
mae = mean_absolute_error(y_test, y_pred_reg)
r2 = r2_score(y_test, y_pred_reg)
rmse = np.sqrt(mse)

print(f'Mean Squared Error (MSE): {mse:.4f}')
print(f'Mean Absolute Error (MAE): {mae:.4f}')
print(f'R² Score: {r2:.4f}')
print(f'Root Mean Squared Error (RMSE): {rmse:.4f}')
```

### MLP Classification

```python
mlp_clf = MLPClassifier(hidden_layer_sizes=(64, 32), activation='relu', solver='adam', max_iter=1000, random_state=42)
mlp_clf.fit(X_train, y_train)
```

```python
y_pred_clf = mlp_clf.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred_clf)
precision = precision_score(y_test, y_pred_clf, average='weighted')
recall = recall_score(y_test, y_pred_clf, average='weighted')
f1 = f1_score(y_test, y_pred_clf, average='weighted')
conf_matrix = confusion_matrix(y_test, y_pred_clf)

print(f'Accuracy: {accuracy:.4f}')
print(f'Precision: {precision:.4f}')
print(f'Recall: {recall:.4f}')
print(f'F1 Score: {f1:.4f}')
print('\nConfusion Matrix:')
print(conf_matrix)
```

```python
print('\nClassification Report:')
print(classification_report(y_test, y_pred_clf))
```

### ANN With Tensorflow

```python
ann = tf.keras.models.Sequential()
```

### Adding the input layer and the first hidden layer

```python
ann.add(tf.keras.layers.Dense(units=4, activation='relu'))
```

### Adding the second hidden layer

```python
ann.add(tf.keras.layers.Dense(units=4, activation='relu'))
```

### Adding the output layer

```python
ann.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

### Compiling the ANN

```python
ann.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy', 'Precision', 'Recall', 'AUC'])
```

### Training the ANN on the Training set

```python
ann.fit(X_train, y_train, batch_size = 64, epochs = 100)
```

```python
y_pred = ann.predict(X_test)
y_pred_ann = np.round(y_pred)
```

## Part 4 - Making the predictions and evaluating the model

```python
accuracy = accuracy_score(y_test, y_pred_ann)
precision = precision_score(y_test, y_pred_ann, average='weighted')
recall = recall_score(y_test, y_pred_ann, average='weighted')
f1 = f1_score(y_test, y_pred_ann, average='weighted')
conf_matrix = confusion_matrix(y_test, y_pred_ann)

print(f'Accuracy: {accuracy:.4f}')
print(f'Precision: {precision:.4f}')
print(f'Recall: {recall:.4f}')
print(f'F1 Score: {f1:.4f}')
print('\nConfusion Matrix:')
print(conf_matrix)
```

### ANN With Pytorch

```python
class MLP(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(MLP, self).__init__()
        self.fc1 = nn.Linear(input_size, hidden_size)
        self.fc2 = nn.Linear(hidden_size, hidden_size)
        self.fc3 = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = torch.sigmoid(self.fc3(x))
        return x
```

```python
X_train_np = X_train.values if isinstance(X_train, pd.DataFrame) else X_train
y_train_np = y_train.values if isinstance(y_train, pd.Series) else y_train

X_test_np = X_test.values if isinstance(X_test, pd.DataFrame) else X_test
y_test_np = y_test.values if isinstance(y_test, pd.Series) else y_test

X_train_tensor = torch.tensor(X_train_np, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train_np, dtype=torch.float32)

X_test_tensor = torch.tensor(X_test_np, dtype=torch.float32)
y_test_tensor = torch.tensor(y_test_np, dtype=torch.float32)
```

```python
input_size = X_train.shape[1]
hidden_size = 64
output_size = 1

model = MLP(input_size, hidden_size, output_size)

criterion = nn.BCELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
num_epochs = 100
batch_size = 32

for epoch in range(num_epochs):
    model.train()

    for i in range(0, X_train_tensor.size(0), batch_size):
        X_batch = X_train_tensor[i:i + batch_size]
        y_batch = y_train_tensor[i:i + batch_size]


        outputs = model(X_batch).squeeze()


        loss = criterion(outputs, y_batch)


        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

    if (epoch + 1) % 10 == 0:
        print(f'Epoch [{epoch + 1}/{num_epochs}], Loss: {loss.item():.4f}')
```

```python
model.eval()

with torch.no_grad():

    y_pred = model(X_test_tensor).squeeze()


y_pred_classes = (y_pred >= 0.5).int().numpy()
y_test_np = y_test_tensor.numpy()


accuracy = accuracy_score(y_test_np, y_pred_classes)
precision = precision_score(y_test_np, y_pred_classes)
recall = recall_score(y_test_np, y_pred_classes)
f1 = f1_score(y_test_np, y_pred_classes)
conf_matrix = confusion_matrix(y_test_np, y_pred_classes)

print(f'Accuracy: {accuracy:.4f}')
print(f'Precision: {precision:.4f}')
print(f'Recall: {recall:.4f}')
print(f'F1-Score: {f1:.4f}')
print('Confusion Matrix:')
print(conf_matrix)
```
